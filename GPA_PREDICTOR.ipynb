## Formulate the problem :
### Description
Cumulative Grade Point Average (CGPA) refers to the overall Grade Point Average (GPA), obtained by dividing
the total Grade Points (GPs) earned in all courses attempted by the total degree-credit hours in all attempted courses.
 You are required to develop a machine learning system to predict final CGPA of a student at the end of fourth
year given GPs of the courses obtained in initial years (up to first, second or third year).


*  The dataset to be used is attached with this file with name The_Grades_Dataset.csv.
*  Model 1: predict final CGPA based on GPs of first year only.
* Model 2: predict final CGPA based on GPs of first two years.
* Model 3: predict final CGPA based on GPs of first three years.

```python
df.replace(np.nan,0, inplace = True)
df.head(10)
```

```python
input = df.drop(["STUDENT ID","100L_1ST_GPA", "100L_1ST_CGPA","100L_1ST_REMARK" ,"100L_2ND_GPA", "100L_2ND_CGPA","100L_2ND_REMARK" ,"200L_1ST_GPA", "200L_1ST_CGPA","200L_1ST_REMARK" ,"200L_2ND_GPA", "200L_2ND_CGPA","200L_2ND_REMARK" ,"300L_1ST_GPA", "300L_1ST_CGPA","300L_1ST_REMARK" ,"300L_2ND_GPA", "300L_2ND_CGPA","300L_2ND_REMARK"], axis=1)
target = df[["100L_1ST_GPA", "100L_1ST_CGPA","100L_1ST_REMARK" ,"100L_2ND_GPA", "100L_2ND_CGPA","100L_2ND_REMARK" ,"200L_1ST_GPA", "200L_1ST_CGPA","200L_1ST_REMARK" ,"200L_2ND_GPA", "200L_2ND_CGPA","200L_2ND_REMARK" ,"300L_1ST_GPA", "300L_1ST_CGPA","300L_1ST_REMARK" ,"300L_2ND_GPA", "300L_2ND_CGPA","300L_2ND_REMARK"]]
```

```python
target.head()
```

```python
pd.unique(df["AFE 101"])
```

```python
grades_enc = { 'A':5.0, 'B':4.0, 'C':3.0, 'D':0.0, 'F':0.0}
```

```python
type(input)
```

```python
for item in input:
      input[item] = input[item].replace(regex=r'^([7-9][0-9]|100)A$', value=5.0)
      input[item] = input[item].replace(regex=r'^[6][0-9]B$', value=4.0)
      input[item] = input[item].replace(regex=r'^[5][0-9]C$', value=3.0)
      input[item] = input[item].replace(regex=r'^[4][1-9]D$', value=2.0)
      input[item] = input[item].replace(regex=r'^[0-4][0-9]F$', value=0.0)
      input[item] = input[item].replace(regex=r'^[0-9]F$', value=0.0)
input.head()
```

```python
# Concating inputs and targets for data insights
df = pd.concat([input, target], axis=1)
with pd.option_context('display.max_rows', None,
                       'display.max_columns', None,
                       'display.precision', 3,
                       ):
    print(df)
```

```python
# Splitting file into different columns based on our model prediction requirement
model1_columns = []
model2_columns = []
model3_columns = []
for item in [input]:
 for i in item:
   if i[3]=='1':
    model1_columns.append(i)
   elif i[3]=='2':
    model2_columns.append(i)
   else:
    model3_columns.append(i)

model2_columns = model1_columns + model2_columns
model3_columns = model2_columns + model3_columns
```

## Prepare the data for Random Forest:

###Model 1: Predict final CGPA based on GPs of first year only.

```python
# Input data
labels = df['300L_2ND_CGPA'].values
features = df[list(model1_columns)].values
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_1 = RandomForestRegressor(n_estimators=20, random_state=0)
rf_1.fit(X_train, y_train)
y_pred = rf_1.predict(X_test)
```

```python
from sklearn import metrics

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print("The score of model for testing set",rf_1.score(X_test, y_test))


```

```python

from sklearn import tree
fn=model1_columns
cn="CGPA"
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=800)
tree.plot_tree(rf_1.estimators_[0],
               feature_names = fn, 
               class_names=cn,
               filled = True);
fig.savefig('rf_individualtree.png')
```
